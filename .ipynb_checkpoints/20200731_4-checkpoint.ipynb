## Image Augmentation

```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array, ImageDataGenerator
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
pangsoo = img_to_array(load_img('/Users/kangd/Data File/pangsoo.png'))
```

```python
pangsoo.dtype
```

```python
pangsoo = pangsoo.astype(np.int32)
```

```python
plt.imshow(pangsoo)
plt.show()
```

```python
pangsoo.shape
```

```python
pangsoo = pangsoo.reshape((1,) + pangsoo.shape)
```

```python
pangsoo.shape
```

```python
datagen = ImageDataGenerator(horizontal_flip=True,
                            vertical_flip=True,
                            zoom_range=0.4,
                            rotation_range=45,
                            shear_range=0.5,
                            width_shift_range=0.2,
                            brightness_range=(0.7, 1.3),
                            fill_mode='nearest')
```

```python
pangsoo_gen = datagen.flow(pangsoo)
```

```python
pangsoo_gen
```

```python
np.array(pangsoo_gen).shape
```

```python
plt.figure(figsize=(10,10))

for i in range(9):
    data = next(pangsoo_gen)
    image = data[0]
    
    plt.subplot(3,3,i+1)
    plt.imshow(np.array(image, dtype=np.int32))
```
