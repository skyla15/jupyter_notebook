### AND / XOR / OR / NAND 

```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
```

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2 # bias
    tmp = np.sum(w*x) + b
    if tmp <= 0 :
        return 0
    else:
        return 1
```

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1    
```

```python
def XOR(x1, x2):
    s1 = NAND(x1,x2)
    s2 = OR(x1,x2)
    y = AND(s1,s2)
    return y
```

### @tf.function

```python
# tf.function : 리턴되는 값이 텐서로 나옴 

@tf.function 
def predict(x):
    w = np.random.randn(1, x.shape[1])
    b = np.random.randn(x.shape[0], 1)
    
    return tf.matmul(x, w, transpose_b = True) + b 
```

###### AND 신경망 구성 

```python
def AND_gate():
    x = np.array([[1,1], [1,0], [0,1], [0,0]]) # 입력ㄱ밧 
    target = np.array([[1], [0], [0], [0]]) 
    w = tf.random.normal([2], 0, 1) # weight 
    b = tf.random.normal([1], 0, 1) # bias
    b_x = 1
    learning_rate = 0.1
    error_sum = 0

    for i in range(2000):

        for j in range(len(x)):
            output = sigmoid(np.sum(x[j] * w) + b_x * b)
            # y _ hat 
            error = target[j][0] - output 
            w = w + x[j] * learning_rate * error 
            b = b + b_x * learning_rate * error
            error_sum += error 
            # sum of difference of ( actual values and output values )

        if i % 200 == 199:
            print(i, error_sum, output)
        
    print('==============출력==============')
    for i in range(len(x)):
        print('x:', x[i], 'y:', target[i], 'output:', sigmoid(np.sum(x[i]*w)+b))

    print('w:', w)
    print('b:', b)
```

###### OR gate

```python
def OR_gate():
    x = np.array([[1,1], [1,0], [0,1], [0,0]]) # 입력값  
    target = np.array([[1], [1], [1], [0]]) 
    w = tf.random.normal([2], 0, 1) # weight 
    b = tf.random.normal([1], 0, 1) # bias
    b_x = 1
    learning_rate = 0.1
    error_sum = 0

    for i in range(2000):

        for j in range(len(x)):
            output = sigmoid(np.sum(x[j] * w) + b_x * b)
            # y _ hat 
            error = target[j][0] - output 
            w = w + x[j] * learning_rate * error 
            b = b + b_x * learning_rate * error
            error_sum += error 
            # sum of difference of ( actual values and output values )

        if i % 200 == 199:
            print(i, error_sum, output)
            
    print('==============출력==============')
    for i in range(len(x)):
        print('x:', x[i], 'y:', target[i], 'output:', sigmoid(np.sum(x[i]*w)+b))

    print('w:', w)
    print('b:', b)
```

###### XOR gate

```python
def XOR_gate():
    x = np.array([[1,1], [1,0], [0,1], [0,0]]) # 입력ㄱ밧 
    target = np.array([[0], [1], [1], [0]]) 
    w = tf.random.normal([2], 0, 1) # weight 
    b = tf.random.normal([1], 0, 1) # bias
    b_x = 1
    learning_rate = 0.1
    error_sum = 0

    for i in range(2000):

        for j in range(len(x)):
            output = sigmoid(np.sum(x[j] * w) + b_x * b)
            # y _ hat 
            error = target[j][0] - output 
            w = w + x[j] * learning_rate * error 
            b = b + b_x * learning_rate * error
            error_sum += error 
            # sum of difference of ( actual values and output values )

        if i % 200 == 199:
            print(i, error_sum, output)
            
    print('==============출력==============')
    for i in range(len(x)):
        print('x:', x[i], 'y:', target[i], 'output:', sigmoid(np.sum(x[i]*w)+b))
        
    print('w:', w)
    print('b:', b)
```
