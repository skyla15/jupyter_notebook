## 9. Apply lambda로 데이터 가공

```python
def get_square(a, b):
    return a ** b
```

```python
get_square(4, 2)
```

```python
titanic['Age'].apply(lambda x: x ** 2)
```

```python
titanic['Name_len'] = titanic['Name'].apply(lambda x: len(x))
titanic.head(2)
```

```python
titanic['Child_Adult'] = titanic['Age'].apply(lambda x: 'Child' 
                                              if x <= 18 else 'Adult')
titanic.head(10)
```

```python
del titanic['Child_Adult']
```

```python
titanic['Age_cat'] = titanic['Age'].apply(lambda x: 'Child' if x <= 18 
                                          else ('Adult' if x <=60 
                                                else 'Elderly'))
```

```python
titanic['Age_cat'].value_counts()
```

```python
titanic.head()
```

```python
titanic.drop(['Cabin', 'Name_len', 'Ticket'], axis=1, inplace=True)
titanic.head()
```

```python
del titanic['Name']
```

```python
titanic.Fare.describe()
```

```python
bin_names = ['low', 'middle', 'high']
titanic['Fare_class'] = pd.cut(titanic['Fare'],     # 데이터 배열
                      bins=3,               # bin_dividers 경계 값 리스트
                      labels=bin_names,     # bin 이름
                      include_lowest=True)
```

```python
titanic.head()
```

```python
titanic.Fare_class.value_counts()
```

```python
df1 = pd.get_dummies(titanic['Pclass'], drop_first=True)
df2 = pd.get_dummies(titanic['Sex'], drop_first=True)
df3 = pd.get_dummies(titanic['Embarked'], drop_first=True)
df4 = pd.get_dummies(titanic['Age_cat'], drop_first=True)
df5 = pd.get_dummies(titanic['Fare_class'], drop_first=True)
```

```python
df = pd.concat([titanic, df1, df2, df3, df4, df5 ],axis=1 )
df.head()
```

```python
df.drop(['Sex', 'Age', 'Fare','Embarked', 'Age_cat', 'Fare_class', 'Family_No'], axis=1, inplace=True)
df.head()
```

```python
df.Survived = df.Survived.astype(int)
```

```python
df.head()
```

```python
y = df['Survived']
y.shape
```

```python
X = df.drop('Survived', axis=1)
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.3, 
                                                    random_state=10)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import warnings 
warnings.filterwarnings('ignore')

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(C=1000, random_state=10)
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
y_pred_prob = lr.predict_proba(X_test)
```

```python
y_pred_prob = lr.predict_proba(X_test)[:, 1]
```

```python
from sklearn import metrics

print(metrics.accuracy_score(y_test, y_pred))
```

```python
y_pred_class = np.where(y_pred_prob >= 0.4, 1, 0)
y_pred_class
```

```python
print(metrics.accuracy_score(y_test, y_pred_class))
```
