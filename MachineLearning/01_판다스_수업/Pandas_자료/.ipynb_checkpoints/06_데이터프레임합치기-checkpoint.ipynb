# 데이터프레임 합치기
- pd.concat( [ 데이터프레임1, 데이터프레임2, ... ] ) 
    - 옵션 
        - join = 'inner'  : 교집합
        - join = outter   : 합집합 
        - ignore_index = True/False

- pd.merge(데이터프레임1, 데이터프레임2)
    - 옵션 : 
        - 기본값 : on = None, how = 'inner'
        - on
            - on = None   : 두 데이터 프레임의 공통으로 속하는 모든 열을 기준 
            - on = "컬럼명" : 컬럼명을 기준으로 
        - how
            - how = 'inner' : 교집합
            - how = 'outer' : 합집합
            - how = 'left' : 왼쪽 데이터프레임의 키(열)은 모두 살리고 이를 기준으로 병합. 
                - right_on : 오른쪽 데이터프레임이 병합 기준 
                - left_on : 왼쪽 데이터프레임의 병합 기준 키값
                    - right_on, left_on 같이 사용 시, 두 키에 맞춰서 두 데이터 프레임 정렬하여 병합 
            - how = 'right' : 오르쪽 데이터프레임의 키(열)값은 모두 살리고 이를 기준으로 병합. 
                - right_on : 오른쪽 데이터프레임이 병합 기준 
                - left_on : 왼쪽 데이터프레임의 병합 기준 키값
                    - right_on, left_on 같이 사용 시, 두 키에 맞춰서 두 데이터 프레임 정렬하여 병합 

- 데이터프레임.join( 데이터프레임 )
    - 기본적으로 merge와 동작 방식 비슷
    - 
    - 옵션
        - 기본값 : on = None, how = 'left' 
        - on = key
        - how = 'left'
        - how = 'right'
        - how = 'inner'

- 인덱스 순으로 데이터 정렬 시 => join 
- 키로 정렬이 가능하도록 정렬이 가능하다면 => merge 
    - aN값을 허용하지 않는 키값들이 각 데이터프레임에 존재할 경우 => merge   
    

```python
# 라이브러리 불러오기
import pandas as pd
```

```python
# IPyhton 디스플레이 설정 변경 
pd.set_option('display.max_columns', 10)                  # 출력할 최대 열의 개수
pd.set_option('display.max_colwidth', 20)                 # 출력할 열의 너비
# pd.set_option('display.max_rows', 1000)                   # 모든 행 보여주기
# pd.set_option('display.max_width', 1000)
```

## `concat()` 함수

```python
# 데이터프레임 만들기
df1 = pd.DataFrame({'a': ['a0', 'a1', 'a2', 'a3'],
                    'b': ['b0', 'b1', 'b2', 'b3'],
                    'c': ['c0', 'c1', 'c2', 'c3']},
                    index=[0, 1, 2, 3])
 
df2 = pd.DataFrame({'a': ['a2', 'a3', 'a4', 'a5'],
                    'b': ['b2', 'b3', 'b4', 'b5'],
                    'c': ['c2', 'c3', 'c4', 'c5'],
                    'd': ['d2', 'd3', 'd4', 'd5']},
                    index=[2, 3, 4, 5])
```

```python
df1
```

```python
df2
```

###### 2개의 데이터프레임을 위 아래 행 방향으로 이어 붙이듯 연결하기 

```python
# 인덱스 겹침, 0 1 2 3 2 3 4 5
```

```python
result1 = pd.concat([df1, df2])
result1
```

```python
# ignore_index=True 옵션 설정하여 기존 index 없애기
```

```python
result2 = pd.concat([df1, df2], ignore_index=True)
result2
```

###### 2개의 데이터프레임을 좌우 열 방향으로 이어 붙이듯 연결하기 

```python
# value가 없으면 NaN(공백)으로 채워짐
result3 = pd.concat([df1, df2], axis=1)
result3
```

```python
# join='inner' 옵션 적용하기 - Inner Join(교집합)
```

```python
result3_in = pd.concat([df1, df2], axis=1, join='inner')
result3_in
```

```python
# join='inner' 옵션 적용하기 - Outer Join(합집합)는 default
```

```python
result3_in = pd.concat([df1, df2], axis=1, join='outer')
result3_in
```

## `merge()` 함수
- pd.merge(df1, df2, options.. )

- 옵션 : 기본값 - on = None, how = 'inner'
    - on
        - on = None   : 두 데이터 프레임의 공통으로 속하는 모든 열을 기준 
        - on = "컬럼명" : 컬럼명을 기준으로 
    - how
        - how = 'inner' : 교집합
        - how = 'outer' : 합집합
        - how = 'left' : 왼쪽 데이터프레임의 키(열)은 모두 살리고 이를 기준으로 병합. 
            - right_on : 오른쪽 데이터프레임이 병합 기준 
            - left_on : 왼쪽 데이터프레임의 병합 기준 키값
                - right_on, left_on 같이 사용 시, 두 키에 맞춰서 두 데이터 프레임 정렬하여 병합 
        - how = 'right' : 오르쪽 데이터프레임의 키(열)값은 모두 살리고 이를 기준으로 병합. 
            - right_on : 오른쪽 데이터프레임이 병합 기준 
            - left_on : 왼쪽 데이터프레임의 병합 기준 키값
                - right_on, left_on 같이 사용 시, 두 키에 맞춰서 두 데이터 프레임 정렬하여 병합 
        

```python
# 주식 데이터를 가져와서 데이터프레임 만들기
```

```python
df1 = pd.read_excel('../../pandas_data/stock price.xlsx')
df1
```

```python
df2 = pd.read_excel('../../pandas_data/stock valuation.xlsx')
df2
```

+ `EPS`(Earning Per Share): 주당순이익
+ `BPS`(Book-value Per Share: 주당순자산가치) = (기업의 현재 순자산) ÷ (총 발행주식수)
+ `PER`(Price Earning Ratio): 주가수익비율 = (주가) ÷ (EPS)
+ `PBR`(Price Book-value Ratio: 주가순자산비율) = (현재주가) ÷ (BPS)


###### 데이터프레임 합치기 - 교집합 ( how = 'inner' )

```python
# default는 Inner Join
# id가 같은 것들끼리 join 
merge_inner = pd.merge(df1, df2)
merge_inner
```

###### 데이터프레임 합치기 - 합집합 ( how = 'outer' )

```python
merge_outer = pd.merge(df1, df2, how='outer')
merge_outer
```

```python
# 데이터프레임 합치기 - 합집합
merge_outer = pd.merge(df1, df2, how='outer', on='id')
merge_outer
```

###### 데이터프레임 합치기 - 왼쪽 데이터프레임 기준, 키 값 분리

```python
# left 기준의 키 = 'name', right 기준의 키 ='name'
# 왼쪽 행의 길이 크기는 그대로, 이 2개의 키에 따라 2개의 데이터가 조인
merge_left = pd.merge(df1, df2, how='left', left_on='stock_name', right_on='name')
merge_left
```

###### left_on, right_on 대신 에, 하나의 키값 'id'를 이용해서 join 

```python
merge_left = pd.merge(df1, df2, how='left', on = 'id')
merge_left
```

```python
# 데이터프레임 합치기 - 오른쪽 데이터프레임 기준, 키 값 분리
```

```python
merge_right = pd.merge(df1, df2, how='right', left_on='stock_name', right_on='name')
merge_right
```

```python
# 불린 인덱싱과 결합하여 원하는 데이터 찾기
```

```python
price_50000 = df1[df1['price'] < 50000]
price_50000
```

```python
value = pd.merge(price_50000, df2)
value
```

## `join()` 함수

```python
df1 = df1.set_index('id')
df2 = df2.set_index('id')
```

```python
print( df1.index )
print( df2.index )
print( [x for x in list(df1.index) if x in list(df2.index)] )
```

###### 데이터프레임 결합(join)

```python
df3 = df1.join(df2)
df3
```

###### 데이터프레임 결합(join) - 교집합

```python
df4 = df1.join(df2, how='inner')
df4
```

###### 데이터프레임 결합(join) - Left Join

```python
df5 = df1.join(df2, how='left')
df5
```

###### 데이터프레임 결합(join) - Right Join

```python
df6 = df1.join(df2, how='right')
df6
```
