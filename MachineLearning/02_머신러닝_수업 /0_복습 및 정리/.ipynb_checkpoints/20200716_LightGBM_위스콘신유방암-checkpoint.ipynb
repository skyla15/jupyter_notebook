```python
# LightGBM   lightgbm LGBMClassifier 
from xgboost import XGBClassifier
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, 
                                                    test_size=0.2)
lgbm_clf = LGBMClassifier(n_estimators=400)
```

```python
lgbm_clf.fit(X_train, y_train)
lgbm_preds = lgbm_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, f1_score, roc_auc_score

def get_clf_eval(y_test , pred):
    confusion = confusion_matrix(y_test, pred)
    accuracy = accuracy_score(y_test , pred)
    precision = precision_score(y_test , pred)
    recall = recall_score(y_test , pred)
    f1 = f1_score(y_test,pred)
    roc_auc = roc_auc_score(y_test, pred)
    print(' ')
    print(confusion)
    print(': {0:.4f}, : {1:.4f}, : {2:.4f},\
    F1: {3:.4f}, AUC:{4:.4f}'.format(accuracy, precision, recall, f1, roc_auc))
```

```python
get_clf_eval(y_test, lgbm_preds)
```

## 

##### LightGBM    ,   min_child_samples (    ), num_leaves (      ), max_depth  . 

```python
xgb_clf1 = XGBClassifier(n_estimators=1000, learning_rate=0.1, max_depth=3, random_state=1)
xgb_clf1.fit(X_train, y_train, eval_metric='auc',
             eval_set=[(X_test, y_test)]) #eval_metric  rmse(), error()
             # eval_set:    . 
xgb_preds1 = xgb_clf1.predict(X_test)
```

```python
get_clf_eval(y_test, xgb_preds1)
```

```python
xgb_clf2 = XGBClassifier(n_estimators=1000, learning_rate=0.1, max_depth=3, random_state=1) #  
xgb_clf2.fit(X_train, y_train, eval_metric='auc', early_stopping_rounds=200,
             eval_set=[(X_test, y_test)]) #eval_metric  rmse(), error()
             # eval_set:    . 
xgb_preds2 = xgb_clf2.predict(X_test) 
```

```python
get_clf_eval(y_test, xgb_preds2)
```

```python
xgb_clf3 = XGBClassifier(n_estimators=1000, learning_rate=0.1, max_depth=3, random_state=1) #  
xgb_clf3.fit(X_train, y_train, early_stopping_rounds=10, eval_metric='auc',
             eval_set=[(X_test, y_test)]) #eval_metric  rmse(), error()
             # eval_set:    . 
xgb_preds3 = xgb_clf3.predict(X_test) 
```

```python
get_clf_eval(y_test, xgb_pred3)
```

```python
from sklearn.model_selection import GridSearchCV

xgb_clf0 = XGBClassifier()
params = {
    'n_estimators':[500, 1000, 2000],
    'learning_rate' : [0.01, 0.05, 0.1],
    'max_depth': [3,4,5,6,7,8,9,10],
    'subsample': [0.5, 0.6, 0.7, 0.8, 0.9, 1.0]
}
grid_cv = GridSearchCV(xgb_clf0, param_grid=params , cv=5, verbose=1, n_jobs=-1)
grid_cv.fit(X_train , y_train)
print('  :\n', grid_cv.best_params_)
print('   : {0:.4f}'.format(grid_cv.best_score_))
```

```python
# GridSearchCV    estimator predict . 
xgb_pred_grid = grid_cv.best_estimator_.predict(X_test)
xgb_accuracy = accuracy_score(y_test, xgb_pred_grid)
print('XGB : {0:.4f}'.format(xgb_accuracy))
```
