```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import numpy as np
from sklearn.base import BaseEstimator

class MyDummyClassifier(BaseEstimator):
    # fit( ) 메소드는 아무것도 학습하지 않음. 
    def fit(self, X , y=None):
        pass
    
    # predict( ) 메소드는 단순히 Sex feature가 1 이면 0 , 그렇지 않으면 1 로 예측함. 
    def predict(self, X):
        pred = np.zeros( ( X.shape[0], 1 ))
        for i in range (X.shape[0]) :
            if X['Sex'].iloc[i] == 1:
                pred[i] = 0
            else :
                pred[i] = 1
        
        return pred

import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Null 처리 함수
def fillna(df):
    df['Age'].fillna(df['Age'].mean(),inplace=True)
    df['Cabin'].fillna('N',inplace=True)
    df['Embarked'].fillna('N',inplace=True)
    df['Fare'].fillna(0,inplace=True)
    return df

# 머신러닝 알고리즘에 불필요한 속성 제거
def drop_features(df):
    df.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
    return df

# 레이블 인코딩 수행. 
def format_features(df):
    df['Cabin'] = df['Cabin'].str[:1]
    features = ['Cabin','Sex','Embarked']
    for feature in features:
        le = LabelEncoder()
        le = le.fit(df[feature])
        df[feature] = le.transform(df[feature])
    return df

# 앞에서 설정한 Data Preprocessing 함수 호출
def transform_features(df):
    df = fillna(df)
    df = drop_features(df)
    df = format_features(df)
    return df

def get_clf_eval(y_test, pred):
    confusion = confusion_matrix(y_test, pred)
    accuracy = accuracy_score(y_test, pred)
    precision = precision_score(y_test, pred)
    recall = recall_score(y_test, pred)
    print('오차 행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}'.format(accuracy , precision ,recall))
    return confusion 


# 위에서 생성한 Dummy Classifier를 이용해 학습/예측/평가 수행.
```

### 정확도 평가 지표와 이진 분류 ML 모델 성능 왜곡 예제 (2) MyFakeClassifier
1) MNIST 세트 - 각 0~9의 값의 픽셀 정보를 갖고 있으나, 7 = True, Other = False로 이진 분류 문제로 변형하여 테스트 
2) 이렇게 불균형한 데이터 세트에 모든 데이터를 False, 즉 0으로 예측하는 classifier를 이용하면 약 90%의 정확도 

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split

from sklearn.base import BaseEstimator
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
import seaborn as sns

class MyFakeClassifier(BaseEstimator):
    def fit(self,X,y):
        pass
    
    # 입력값으로 들어오는 X 데이터 셋의 크기만큼 모두 0값으로 만들어서 반환
    def predict(self,X):
        return np.zeros( (len(X), 1) , dtype=bool)

# 사이킷런의 내장 데이터 셋인 load_digits( )를 이용하여 MNIST 데이터 로딩
digits = load_digits()


print(digits.data)
print("> digits.data.shape:", digits.data.shape)
print('>', digits.target)
print("> digits.target.shape:", digits.target.shape)


# digits번호가 7번이면 True이고 이를 astype(int)로 1로 변환, 7번이 아니면 False이고 0으로 변환. 
y = (digits.target == 7).astype(int)


X_train, X_test, y_train, y_test = train_test_split( digits.data, y, random_state=11)


# 불균형한 레이블 데이터 분포도 확인. 
print('======= 레이블 테스트 세트 크기 :', y_test.shape)
print('> 테스트 세트 레이블 0 과 1의 분포도')
print(pd.Series(y_test).value_counts())
print('> 트레인 세트 레이블 0 과 1의 분포도')
print(pd.Series(y_train).value_counts())


# Dummy Classifier로 학습/예측/정확도 평가
fakeclf = MyFakeClassifier()
fakeclf.fit(X_train , y_train)
fakepred = fakeclf.predict(X_test)
print('======= 모든 예측을 0으로 하여도 정확도는:{:.3f}'.format(accuracy_score(y_test , fakepred)))

sns.distplot(pd.DataFrame(y), kde = False)
```

### 이진 분류에 적합하지 않은 지표 : Accuracy

- Accuracy  ( 정확도 ) : 전체 데이터 중 실제 값과 동일 한 것 ( TN + TP / TN + TP + FN + FP ) 
    - 불균형한 이진 트리에서는, TN( 혹은 TP ) 에 값이 대부분 속하기 때문에, 모델 성능 측정이 왜곡될 수 있음
    - 특히 Negative로 예측할 때 정확도가 높아지는 경향이 있음.

### 이진분류에 적합한 지표 : Precision, Recall 

- Precision ( 정밀도 ) : 암환자 예측 중에 진짜 암환자       ( TP / FP + TP )
    - False Positive를 낮추는 게 목적 !!
        - Ex - 실제 스팸 메일이 아닌 것(Negative)를 스팸 메일이라 판단(False Positive)하는 경우 등 

- Recall    ( 재현율 ) : 진짜 암환자 중에 암환자 예측       ( TP / FN + TP ) 
    - Sensitivity
    - False Negative를 낮추는 게 목적 !!
        - Ex - 금융사기 : 실제 사기를 가짜라 판단할 경우 ) 
        - Ex - 암환자 : 실제 암환자를 암환자가 아니라고 판단할 경우, 목숨의 제 등 매우 리스크가 큼 

- Specificity ( 특이성): 실제 암환자 아닌 사람 중에 실제 암환자 아닌 사람 예측 ( TN / TN + FP )

```python
from sklearn.metrics import accuracy_score, precision_score , recall_score , confusion_matrix

def get_clf_eval(y_test , pred):
    confusion = confusion_matrix( y_test, pred)
    accuracy = accuracy_score(y_test , pred)
    precision = precision_score(y_test , pred)
    recall = recall_score(y_test , pred)
    print('오차 행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}'.format(accuracy , precision ,recall))
    return confusion

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression

# 원본 데이터를 재로딩, 데이터 가공, 학습데이터/테스트 데이터 분할. 
titanic_df = pd.read_csv('./titanic_train.csv')
y_titanic_df = titanic_df['Survived']
X_titanic_df= titanic_df.drop('Survived', axis=1)
X_titanic_df = transform_features(X_titanic_df)

X_train, X_test, y_train, y_test = train_test_split(X_titanic_df, y_titanic_df, \
                                                    test_size=0.20, random_state=11)

lr_clf = LogisticRegression()

lr_clf.fit(X_train , y_train)
pred = lr_clf.predict(X_test)
get_clf_eval(y_test , pred)
```

### predict_proba() 메소드 
- 개별 데이터별로 예측 확률을 반환 메소드
- 학습이 완료된 Classifier객체에서 호출이 가능
- 반환 결과 
    - predict() : 예측 결과 클래스 값
    - proba() : 예측 확률 
- predict_proba 후 반환되는 ndarray 
    - 테스트 세트에서 각 칼럼에 대한 결과값이 나올 확률 반환  
    => predict() 함수는 predict_proba() 호출을 통해 나오는 값들에 대해, 임계치 값 조절을 통해 결과 예측. 

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression

# 원본 데이터를 재로딩, 데이터 가공, 학습데이터/테스트 데이터 분할. 
titanic_df = pd.read_csv('./titanic_train.csv')
y_titanic_df = titanic_df['Survived']
X_titanic_df= titanic_df.drop('Survived', axis=1)
X_titanic_df = transform_features(X_titanic_df)


# 학습/검증 데이터 분리 
X_train, X_test, y_train, y_test = train_test_split(X_titanic_df, y_titanic_df, \
                                                    test_size=0.20, random_state=11)

# 분류기 학습 
lr_clf = LogisticRegression()
lr_clf.fit(X_train , y_train)


# 학습된 분류기를 통해 proba 계산 
pred_proba = lr_clf.predict_proba(X_test)
pred  = lr_clf.predict(X_test)
print('pred_proba()결과 Shape : {0}'.format(pred_proba.shape))
```

### Binarizer과 threshold, 그리고  precision, recall, accuracy 
- Binarizer : threshold값보다 값이 작으면 0, 크면 1
     - threshold와 Binarizer를 통해 predict를 통해 나온 값들을 TN, FN, TP, FP 분류하여 오차행렬 생성 

```python
from sklearn.metrics import accuracy_score, precision_score , recall_score , confusion_matrix
from sklearn.preprocessing import Binarizer
from sklearn.linear_model import LogisticRegression
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns 
import matplotlib.font_manager as fm 
from matplotlib import rc

font_list = [f.name for f in fm.fontManager.ttflist if 'Gothic' in f.name]
print(font_list)
rc('font', family = 'Nanum Gothic')


titanic_df = pd.read_csv('./titanic_train.csv')
y_titanic_df = titanic_df['Survived']
X_titanic_df = titanic_df.drop('Survived', axis=1)

X_titanic_df = transform_features(X_titanic_df)

X_train, X_test, y_train, y_test = train_test_split(X_titanic_df, y_titanic_df, \
                                                    test_size = 0.2, random_state = 11)

# 학습 
lr_clr = LogisticRegression()
lr_clr.fit(X_train, y_train)

accuracy_list = []
precision_list = []
recall_list = []
threshold_list = []
TP_list = []
FN_list = []


for i in range(4):
    #Binarizer의 threshold 설정값. 분류 결정 임곗값임.  
    custom_threshold = i/4
    threshold_list.append(custom_threshold)
    
    pred_proba = lr_clr.predict_proba(X_test)
    
    # predict_proba( ) 반환값의 두번째 컬럼 , 즉 Positive 클래스 컬럼 하나만 추출하여 Binarizer를 적용
    pred_proba_1 = pred_proba[:,1].reshape(-1,1)

    # threshold보다 높은 값을 갖는 행들을 Positive로 해석 
    binarizer = Binarizer(threshold = custom_threshold)
    binarizer.fit(pred_proba_1) 
    custom_predict = binarizer.transform(pred_proba_1)
    
    
    # threshold가 커질수록 precision은 커지고(FP 작아지고), recall 작아짐(FN 커짐)
    clf_eval = get_clf_eval(y_test, custom_predict)
    TN, FP, FN, TP = clf_eval[0][0], clf_eval[0][1], clf_eval[1][0], clf_eval[1][1]
    a, b = FN, TP
    TP_list.append(a/100)
    FN_list.append(b/100)
    
    precision_list.append(TP / ( FP + TP ))
    recall_list.append(TP / ( FN + TP ))
    accuracy_list.append((TP + TN) / (TP + TN + FN + FP))


plt.figure(figsize = (15,8))
plt.plot(threshold_list, precision_list, label = 'precision')
plt.plot(threshold_list, recall_list, label = 'recall')
plt.plot(threshold_list, accuracy_list, linestyle=':', label = 'accuracy')
plt.plot(threshold_list, TP_list, label = 'TP', linestyle='--')
plt.plot(threshold_list, FN_list, label = 'FN', linestyle='-.')
plt.legend()
plt.xlabel('Threshold value'); plt.ylabel('Precision and Recall value')


```

- Precision과 Recall값이 비슷할 때만 높은 값을 갖음 
    - 따라서, precision 혹은 Recall에 균등하지 않은 중요도를 주고 싶을 때는 다른 metrics 사용
        - PR curve / ROC Curve
- PR Curve 
    - 임계값에 따라, FP가 커지면, TP 
        - Precision : 임계값이 커질수록 커짐 ( FP 감소 ) 
        - Recall : 임계값이 커질수록 작아짐 ( FN 증가 ) 
        - 하나의 지표값을 확인하기 위해 사용
    - AP 
        - PR 커브의 아래부터 Baseline까지 넓이
- ROC Curve
    - 임곗값이 커짐에 따라 FN은 커지게 되고, FP는 감소
        - 즉, FN이 커짐에 따른 TP를 확인하는 것 
        - 암에 걸렸지만 암이 아니라고 판정 받는 사람들과 암에 걸렸고 실제 암에 걸린 사람들의 관계를 보여주는 곡선 
    - 잘못된 판정(False Negative)에 예민한 문제를 해결할 때 써야함.
    - 따라서 잘못된 판정에 대한 비율이 1 증가할 때(임곗값을 올릴때), 제대로 된 판정이 10이 증가하면

```python
from sklearn.metrics import accuracy_score, precision_score , recall_score, f1_score
from sklearn.metrics import confusion_matrix, precision_recall_curve
from sklearn.linear_model import LogisticRegression 
from sklearn.preprocessing import Binarizer 
from sklearn.model_selection import train_test_split

from functools import reduce
import pandas as pd


def draw_curves(confusion_matrix_list, f1_score_list):
    TN_list = []
    FP_list = []
    FN_list = []
    TP_list = []
    accuracy_list = []
    precision_list = []
    threshold_list = []
    recall_list = []
    for threshold, elements in confusion_matrix_list.items():
        TN, FP, FN, TP = elements
        accuracy = (TN + TP) / (TN + FP + FN + TP)
        precision = TP / (TP + FP)
        recall = TP / (FN + TP)
        a, b, c, d = TN, FP, FN, TP
        TN_list.append(a / 50)
        FP_list.append(b / 50)
        FN_list.append(c / 50)
        TP_list.append(d / 50)
        accuracy_list.append(accuracy)
        precision_list.append(precision)
        recall_list.append(recall)
        threshold_list.append(threshold)
        
    plt.figure(figsize = (10,10))
    plt.plot(threshold_list, accuracy_list, linestyle = ':', label = 'accuracy')
    plt.plot(threshold_list, precision_list, linestyle = '-', linewidth = 2, label = 'precision')
    plt.plot(threshold_list, recall_list, linestyle = '-', linewidth = 2, label = 'recall')
#     plt.plot(threshold_list, FN_list, linestyle = '--', label = 'FN')
#     plt.plot(threshold_list, TP_list, linestyle = '-.', label = 'TP') 
    plt.plot(threshold_list, f1_score_list, linestyle = '-', linewidth=2, label = 'F1_Score' )
    plt.legend()
    
    
    plt.figure(figsize = (10,10))
    plt.plot(recall_list, precision_list)
    plt.title('PR Curve')
    plt.xlabel('Recall'); plt.ylabel('Precision')

     

# 오차 행렬 내부 원소들 값을 1차원 배열로 넘겨줌 
def get_confusion_elements(confusion):
    return reduce(lambda x,y : x + y, confusion)


# 오차행렬과, f1스코어 리스트 반환 
def get_eval_by_threshold(y_test, pred_proba, n = 30):
    confusion_matrix_list = dict()
    f1_score_list = []
    for i in range(1,n):
        my_threshold = i/40
        binarizer = Binarizer(my_threshold)
        binarizer.fit(pred_proba)
        my_predict = binarizer.transform(pred_proba)
#         print('Threshold', my_threshold)
        confusion, f1 = get_clf_score(y_test, my_predict)
        confusion_elements = get_confusion_elements(confusion.tolist())
#         print('TN, FP, FN, TP : ', confusion_elements)
#         print('오차 행렬')
#         print(confusion)
#         print('=========================')
        
        f1_score_list.append(f1)
        confusion_matrix_list[my_threshold] = confusion_elements
    return confusion_matrix_list, f1_score_list
        

        
def get_clf_score(y_test, pred):
    confusion = confusion_matrix(y_test, pred)
    accuracy = accuracy_score(y_test, pred)
    precision = precision_score(y_test, pred)
    recall = recall_score(y_test, pred)
    f1 = f1_score(y_test, pred)
#     print('정확도: {:.4f}'.format(accuracy))
#     print('정밀도: {:.4f}'.format(precision))
#     print('재현율: {:.4f}'.format(recall))
#     print('F1 스코어: {:.4f}'.format(f1))
    return confusion, f1 



df_titanic = pd.read_csv('./titanic_train.csv')
X_titanic = df_titanic.drop('Survived', axis = 1)
X_titanic = transform_features(X_titanic)
y_titanic = df_titanic['Survived']

X_train, X_test, y_train, y_test = train_test_split(X_titanic, y_titanic, \
                                                    test_size = 0.2, random_state = 11)

lr_clf = LogisticRegression()
lr_clf.fit(X_train, y_train)

pred_proba = lr_clf.predict_proba(X_test)
pred = lr_clf.predict(X_test)




# 그림을 그리기위해 각 threshold와 오차행렬을 가져옴
confusion_matrix_list, f1_score_list = get_eval_by_threshold(y_test, pred_proba[:,1].reshape(-1,1))
draw_curves(confusion_matrix_list, f1_score_list)
```

#### ROC - Curve vs  PR-Curve 
PR curve whenever the positive class is rare or when you care more about the false positives than the false negatives. Otherwise, use the ROC curve.
- PR Curve 
    - 임계값에 따라, FP가 커지면, TP 
        - Precision : 임계값이 커질수록 커짐 ( FP 감소 ) 
        - Recall : 임계값이 커질수록 작아짐 ( FN 증가 ) 
        - 하나의 지표값을 확인하기 위해 사용
    - AP 
        - PR 커브의 아래부터 Baseline까지 넓이

- ROC Curve : Threshold 크기 변화에 따른 FPR 증가에 따른 TPR 과의 관계 비교 
    - 임곗값이 커짐에 따라 FN은 커지게 되고, FP는 감소, 따라서 TPR은 증가하게 됨 
        - 즉, FN이 커짐에 따른 TPR를 확인하는 것 
        - 암에 걸렸지만 암이 아니라고 판정 받는 사람들이 증감함에 따라 
            - 암에 걸렸다고 판단받은 사람들 중에 실제 암에 걸린 사람들의 관계를 보여주는 곡선
    - 언제 사용?
        - 잘못된 판정(False Negative)에 예민한 문제를 해결할 때 써야함.
        - 따라서 잘못된 판정에 대한 비율이 1 증가할 때(임곗값을 올릴때), 제대로 된 판정이 10이 증가하면 좋은거 

- FPR-TPR
    - TPR = Recall(재현율, 민감도) = TP / ( FN + TP )
    - TNR = 특이성 = TN / (FP + TN)
    - FPR = FP / (FP + TN ) = 1 - TNR
