```python
import numpy as np
import seaborn as sns
import pandas as pd 

import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

from matplotlib import rc
```

```python
%matplotlib inline 
```

```python
# 샘플 데이터 
titanic = sns.load_dataset('titanic')
```

### 분포 시각화 - 카테고리형 자료 

```python
# boxplot, countplot, barplot, pointplot, distplot

plt.figure(figsize=(20,30))
plt.subplot(331)
sns.boxplot(x = titanic['class'], y = titanic['age'], hue=titanic['survived'])
plt.subplot(332)
sns.boxplot(x = titanic['class'], y = titanic['parch'], hue=titanic['survived'])
plt.subplot(333)
sns.boxplot(x = titanic['class'], y = titanic['fare'], hue=titanic['survived'])

# 죽은 남자와 여자들의 크기는 얼마나 될까 
plt.subplot(334)
sns.countplot(x = titanic['sex'], hue=titanic['survived'])
# 탑승하고 있던 클래스 숫자
plt.subplot(335)
sns.countplot(x = titanic['class'], hue=titanic['survived'])
# 각 항에서 탑승한 사람들
plt.subplot(336)
sns.countplot(x = titanic['embarked'], hue = titanic['survived'])

# 나이 분포는 어떻게 되나?
plt.subplot(337)
sns.distplot(titanic['age'], hist = True)
# 죽은 사람들과 생존한 사람들의 나이 분포는 어떻게 되나?
plt.subplot(338)
sns.distplot(titanic[titanic['survived'] == 0 ]['age'], hist = True, label = 'dead')
sns.distplot(titanic[titanic['survived'] == 1 ]['age'], hist = True, label = 'survived')
# 죽은 사람들과 생존한 사람들이 질분한 배 운임은 어떻게 될까
plt.subplot(339)
sns.distplot(titanic[titanic['survived'] == 0 ]['fare'], hist = True, label = 'dead')
sns.distplot(titanic[titanic['survived'] == 1 ]['fare'], hist = True, label = 'survived')
```

```python
# 카운트플롯과 바플롯 
plt.figure(figsize = (15,15))

plt.subplot(221)
sns.countplot(x = titanic['sex'], hue = titanic['survived'])
plt.subplot(222)
sns.countplot(x = titanic['class'], hue = titanic['survived'])

plt.subplot(223)
sns.barplot(x = titanic['sex'], y = titanic['age'], hue = titanic['survived'] )
plt.subplot(224)
sns.barplot(x = titanic['class'], y = titanic['age'], hue = titanic['survived'] )
```
