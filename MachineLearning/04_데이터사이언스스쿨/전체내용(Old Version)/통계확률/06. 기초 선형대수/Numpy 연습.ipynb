### meshgrid

```python
x = np.arange(3)
x

y = np.arange(5)
y
```

```python
X, Y = np.meshgrid(x, y)
```

```python
for x, y in zip(X, Y):
    print(x, y)
```

```python
[list(zip(x, y)) for x, y in zip(X, Y)]
```
